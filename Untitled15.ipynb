**Classification Decision Tree vs. Multiple Linear Regression**:
- **Classification Decision Tree**: Used for predicting categories (e.g., Yes/No, Cat/Dog) by making a series of decisions (like a flowchart).
- **Multiple Linear Regression**: Used for predicting numerical values (e.g., price, weight) by finding the best line through data points.

**Metrics and Real-World Applications**:
1. **Accuracy**: Suitable for situations where both positives and negatives matter equally, like **quality control** in manufacturing.
2. **Sensitivity**: Important when missing true positives is costly, such as in **disease screening** (e.g., cancer).
3. **Specificity**: Crucial when false positives could cause harm, like **testing for a rare disease**.
4. **Precision**: Needed to minimize false positives, e.g., **spam detection** in emails, to ensure flagged items are correct.

```python
# Q6 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
import numpy as np
import graphviz as gv

# Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Step 1: Drop unnecessary columns
ab_reduced = ab.drop(columns=['Weight_oz', 'Width', 'Height'])  # This step removes unnecessary columns that are not needed for the analysis, helping to simplify the dataset.

# Step 2: Drop rows with any NaN values in the remaining columns
ab_reduced_noNaN = ab_reduced.dropna()

# Step 3: Convert columns to appropriate data types
ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

# Step 4: Create an 80/20 split for training and testing datasets
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# Report the number of observations in the training and test datasets
print(f"Number of observations in training set: {len(ab_reduced_noNaN_train)}")
print(f"Number of observations in test set: {len(ab_reduced_noNaN_test)}")

# Tell a ChatBot about fitting a DecisionTreeClassifier model
# ChatBot help: We are about to fit a "scikit-learn" DecisionTreeClassifier model. The two steps are as follows:
# Step 1: Prepare the target variable (y) and features (X) for model training.
# Step 2: Train the classifier using the training dataset.

# Step 5: Prepare the target variable and features for the model
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']
X = ab_reduced_noNaN_train[['List Price']]

# Step 6: Train the DecisionTreeClassifier
clf = tree.DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

# Step 7: Visualize the decision tree
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)

# Step 8: Train a second DecisionTreeClassifier using additional features
X2 = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]
clf2 = tree.DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X2, y)

# Step 9: Visualize the second decision tree
tree.plot_tree(clf2, feature_names=['NumPages', 'Thick', 'List Price'], class_names=['Paperback', 'Hardcover'], filled=True)

# Step 10: Evaluate both models using the test set
# Prepare test set target variable and features
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']
X_test_1 = ab_reduced_noNaN_test[['List Price']]
X_test_2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]

# Predictions for clf and clf2
y_pred_clf = clf.predict(X_test_1)
y_pred_clf2 = clf2.predict(X_test_2)

# Confusion matrix and metrics for clf
conf_matrix_clf = confusion_matrix(y_test, y_pred_clf)
TP_clf, TN_clf, FP_clf, FN_clf = conf_matrix_clf[1, 1], conf_matrix_clf[0, 0], conf_matrix_clf[0, 1], conf_matrix_clf[1, 0]
sensitivity_clf = np.round(TP_clf / (TP_clf + FN_clf), 3)
specificity_clf = np.round(TN_clf / (TN_clf + FP_clf), 3)
accuracy_clf = np.round(accuracy_score(y_test, y_pred_clf), 3)

print(f"Confusion Matrix for clf:\n{conf_matrix_clf}")
print(f"Sensitivity for clf: {sensitivity_clf}")
print(f"Specificity for clf: {specificity_clf}")
print(f"Accuracy for clf: {accuracy_clf}")

# Confusion matrix and metrics for clf2
conf_matrix_clf2 = confusion_matrix(y_test, y_pred_clf2)
TP_clf2, TN_clf2, FP_clf2, FN_clf2 = conf_matrix_clf2[1, 1], conf_matrix_clf2[0, 0], conf_matrix_clf2[0, 1], conf_matrix_clf2[1, 0]
sensitivity_clf2 = np.round(TP_clf2 / (TP_clf2 + FN_clf2), 3)
specificity_clf2 = np.round(TN_clf2 / (TN_clf2 + FP_clf2), 3)
accuracy_clf2 = np.round(accuracy_score(y_test, y_pred_clf2), 3)

print(f"Confusion Matrix for clf2:\n{conf_matrix_clf2}")
print(f"Sensitivity for clf2: {sensitivity_clf2}")
print(f"Specificity for clf2: {specificity_clf2}")
print(f"Accuracy for clf2: {accuracy_clf2}")

# Summary of the session:
# In this script, we start by loading the Amazon Books dataset and proceed with data preprocessing. First, we remove unnecessary columns ('Weight_oz', 'Width', 'Height') to simplify the dataset. Next, we drop any rows containing NaN values to ensure the data is complete for subsequent analysis. We then convert 'Pub year' and 'NumPages' columns to integer type to maintain consistency for numerical operations, while 'Hard_or_Paper' is converted to a categorical type for memory optimization and categorical analysis. We then create an 80/20 split of the cleaned dataset for training and testing purposes, ensuring reproducibility by setting a random seed. After splitting the data, we proceed to train a DecisionTreeClassifier model using only the 'List Price' variable to predict whether a book is a hardcover or paperback, with a maximum depth of 2. We then visualize the decision tree to explain the predictions made based on the 'List Price' variable. Next, we train a second DecisionTreeClassifier model using additional features ('NumPages', 'Thick', 'List Price') with a maximum depth of 4. We also visualize the second decision tree to understand the more complex predictions made based on multiple features. Finally, we evaluate both models using the test set by creating confusion matrices and calculating sensitivity, specificity, and accuracy for each model.
```

We discussed the differences between interpreting coefficients in linear regression and feature importances in decision trees. In linear regression, coefficients quantify how much the outcome changes for a unit change in each feature. In decision trees, feature importance indicates how valuable each feature is for improving the models accuracy, but it doesn't provide a direct numerical effect on the outcome. The key difference is that linear regression provides a direct measure of effect, while decision trees assess overall contribution.
