Problem introduction 
In the early 1920s a experiment was conducted between Ronald Fisher and Dr. Muriel Bristol. to test if there a difference in taste of milk was poured before or after the tea. In this project we will conduct similiar experiments using a sample of 80 stat130 stusdents to determinate if they could taste the difference. 80 out of 49 students were able to correctly identify the pouring order. 

Relationship to original experiments 
The main difference between the two experiments is that fisher's experiment is very personize and the stat student one is not as much and futhermore we have a way larger sample size. 

Hypotheses 
In theory that if they are guessing the spread should be around 40 on guessing right or wrong from the data we could see that it is more close to 50 people telling a difference. 


Methodology Code and Explanation
WE will us a binomial test to determine the probability of gettign 49 or more correct out of 80. if we assume that people are randomely guessing the puring order is 0.5. 

```python
#Methodology Code and Explanation
 


import numpy as np
import scipy.stats as stats

# Set seed for reproducibility
np.random.seed(42)

# Given Data
n = 80  # Total students
x = 49  # Correct responses
p = 0.5  # Null hypothesis probability

# Performing the Binomial test
p_value = stats.binom_test(x, n, p, alternative='greater')
print(f"P-value: {p_value:.4f}")

# Simulating the sampling distribution
simulations = 10000
sample_distribution = np.random.binomial(n, p, simulations)

# Calculating the proportion of simulations where the number of correct answers is 49 or more
prob_49_or_more = np.sum(sample_distribution >= x) / simulations
print(f"Simulated P-value: {prob_49_or_more:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
n_students = 80  # Total number of students
n_trials = 10000  # Number of simulations
correct_identifications = 49  # Number of correct identifications

# Perform simulations for the null hypothesis (p = 0.5)
simulated_correct_counts = np.random.binomial(n_students, 0.5, n_trials)

# Visualization of the simulation distribution
plt.figure(figsize=(10, 6))
plt.hist(simulated_correct_counts, bins=30, alpha=0.7, color='red', edgecolor='black', density=True)
plt.axvline(correct_identifications, color='green', linestyle='dashed', linewidth=2, 
            label='Observed Correct Identifications (49)')
plt.title('Distribution of Simulated Correct Identifications under Null Hypothesis')
plt.xlabel('Number of Correct Identifications')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```
